```python
import os
import cv2
import csv
import numpy as np
import random
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
img = cv2.imread('./data/img/center_2017_02_17_11_45_53_453.jpg')
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
flipped_img = cv2.flip(rgb_img, 1)




plt.imshow(rgb_img)
plt.show()

plt.imshow(flipped_img)
plt.show()

```

```python
measurements = []
with open('./data/driving_log.csv') as csvfile:
    reader = csv.reader(csvfile)
    for line in reader:
        correction = 0.2
        measurement = float(line[3])
        measurements.append(measurement)
        measurements.append(measurement+correction)
        measurements.append(measurement-correction)
        measurements.append(measurement * -1.0)
        measurements.append((measurement+correction) * -1.0)
        measurements.append((measurement-correction) * -1.0)

plt.hist(measurements, bins=100)
plt.show()        
```

```python
for pct in np.arange(0.4, 1.3, .1):
    image = cv2.imread('./data/IMG/center_2017_02_17_10_34_38_221.jpg')
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    hsv[:,:,2] = hsv[:,:,2] * pct
    image_brightness = cv2.cvtColor(hsv, cv2.COLOR_HSV2RGB)
    plt.imshow(image_brightness)
    plt.title('pct_change: ' + str(pct))
    plt.show()
```

```python
final_images = []
for i in range(10):
    path = './data/IMG/'
    rndm_img = random.choice(os.listdir('./data/IMG/'))
    image = path + rndm_img
    image = cv2.imread(image)
    change_pct = random.uniform(0.4, 1.2)
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    hsv[:,:,2] = hsv[:,:,2] * change_pct
    image_brightness = cv2.cvtColor(hsv, cv2.COLOR_HSV2RGB)
    cropped = image_brightness[70:135, 30:290]
    plt.imshow(cropped)
    plt.show()

    
```
